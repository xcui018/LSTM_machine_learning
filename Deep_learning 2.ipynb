```python
x_train = input_list[0:8001]
x2_train = input_month_list[0:8001]
#x3_train = np.array(input_multiplier_list[0:400000])
#x4_train = np.array(input_nzed_list[0:400000])

y_train = output_list[0:8001]


x_test = input_list[8001:]
x2_test = input_month_list[8001:]
#x3_test = np.array(input_multiplier_list[400000:])
#x4_test = np.array(input_nzed_list[400000:])

y_test = output_list[8001:]


a1_train = np.array(x_train)
a2_train = np.array(x2_train)
#a3_train = np.array(x3_train)
#a4_train = np.array(x4_train)

a1_test = np.array(x_test)
a2_test = np.array(x2_test)
#a3_test = np.array(x3_test)
#a4_test = np.array(x4_test)
#a3 = np.concatenate((a1,a2),axis = 1)


```

```python
aa1_train = np.reshape(a1_train, (a1_train.shape[0], a1_train.shape[1], 1))
aa2_train  = np.reshape(a2_train, (a2_train.shape[0], a2_train.shape[1], 1))
#aa3_train = np.reshape(a3_train, (a3_train.shape[0], a3_train.shape[1], 1))
#aa4_train = np.reshape(a4_train, (a4_train.shape[0], a4_train.shape[1], 1))
train_input = np.dstack((aa1_train,aa2_train))

labels = np.array(y_train)


```

```python
aa1_test = np.reshape(a1_test, (a1_test.shape[0], a1_test.shape[1], 1))
aa2_test  = np.reshape(a2_test, (a2_test.shape[0], a2_test.shape[1], 1))
#aa3_test  = np.reshape(a3_test, (a3_test.shape[0], a3_test.shape[1], 1))
#aa4_test  = np.reshape(a4_test, (a4_test.shape[0], a4_test.shape[1], 1))
test_input = np.dstack((aa1_test,aa2_test))
```

```python
model = Sequential()
model.add(LSTM(units=10, return_sequences=True, input_shape=(train_input.shape[1], 2)))
model.add(Dropout(0.2))
model.add(LSTM(units=10, return_sequences=True))  
model.add(Dropout(0.2))

#model.add(LSTM(units=10, return_sequences=True))  
#model.add(Dropout(0.2))

model.add(LSTM(units=10))  
model.add(Dropout(0.2))  

#model.add(Dense(units = 60),)

model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mse')

```

```python
model.fit(train_input, labels, epochs = 100, batch_size = 10)
```

```python
comparison_2inputs_alpha = pd.DataFrame({'predict':inver_predict,'Origin':inver_y_test})
comparison_2inputs_alpha = comparison_2inputs_alpha.round(2)
comparison_2inputs_alpha['diff'] = comparison_2inputs_alpha['predict']-comparison_2inputs_alpha['Origin']
comparison_2inputs_alpha['sq_diff'] = comparison_2inputs_alpha['diff']*comparison_2inputs_alpha['diff']
comparison_2inputs_alpha['abs_diff'] = abs(comparison_2inputs_alpha['diff'])
comparison_2inputs_alpha['diff_percent'] = comparison_2inputs_alpha['diff']/comparison_2inputs_alpha['Origin']
comparison_2inputs_alpha['abs_diff_percent'] = abs(comparison_2inputs_alpha['diff_percent'])
```

```python
new_p = []
for i,row in enumerate(comparison_2inputs_alpha.itertuples()):
    if str(row.abs_diff_percent)=='inf':
        new_p.append(1)
    elif row.abs_diff_percent>1:
        new_p.append(1)
    else:
        new_p.append(row.abs_diff_percent)

comparison_2inputs_alpha['abs_diff_percent'] = new_p
```
